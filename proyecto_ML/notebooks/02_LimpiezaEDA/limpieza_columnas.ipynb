```python
def extract_info(json_string, field='name'):
    """
    Extrae información de una columna en formato JSON
    """
    if json_string is None or (isinstance(json_string, float) and pd.isna(json_string)):
        return []
    
    json_string = str(json_string)
    
    if json_string == '' or json_string == '[]' or json_string == 'nan' or json_string == 'None':
        return []
    
    try:
        data = ast.literal_eval(json_string)
        
        # Si es un diccionario único
        if isinstance(data, dict):
            return [data.get(field, '')] if data.get(field) else []
        
        # Si es una lista de diccionarios
        if isinstance(data, list):
            values = [item.get(field, '') for item in data if isinstance(item, dict) and field in item]
            return values
        
        return []
    except:
        return []

# Columnas JSON a descomprimir
columnas_json = {
    'production_companies': 'name',
    'production_countries': 'name',
    'keywords': 'name',
    'cast': 'name',
    'crew': 'name',
    'genres': 'name',
    'similar_movies': 'title'
}

for col, field in columnas_json.items():
    if col in df.columns:
        nueva_col = f"{col}_names"
        df[nueva_col] = df[col].apply(lambda x: extract_info(x, field=field))
        
        no_vacios = df[nueva_col].apply(len).sum()
        promedio = df[nueva_col].apply(len).mean()
        
        print(f"✓ {col} → {nueva_col}")
        print(f"  Total: {no_vacios} | Promedio: {promedio:.2f}")

print(" CREamos la variable is_franchise")

# Crear variable booleana para belongs_to_collection
df['is_franchise'] = df['belongs_to_collection'].apply(
    lambda x: 0 if x is None or str(x) == 'None' or str(x) == '[]' else 1
)

print(f" is_franchise creada")
print(f"  En franquicia: {df['is_franchise'].sum()} ({df['is_franchise'].mean()*100:.1f}%)")
print(f"  Independientes: {(df['is_franchise']==0).sum()} ({(df['is_franchise']==0).mean()*100:.1f}%)")

# Columnas finales que queremos mantener (sin las JSON originales, son _names)
columnas_finales = [
    'title', 'original_language', 'release_date', 
    'budget', 'revenue', 'runtime', 
    'popularity', 'vote_average', 'vote_count',
    'is_franchise',
    'genres_names', 'production_companies_names', 'production_countries_names',
    'keywords_names', 'cast_names', 'crew_names', 'similar_movies_names'
]

# Crear df_total solo con las columnas finales
df_total = df[columnas_finales].copy()

print(f" df_total creado")
print(f"  Shape: {df_total.shape}")
print(f"  Columnas: {len(df_total.columns)}")


print(f"\nColumnas en df_total:")
for i, col in enumerate(df_total.columns, 1):
    print(f"{i:2d}. {col}")

# Verificar ejemplo
print( "Verificamos con un ejemplo :")
print(df_total[['title', 'revenue', 'is_franchise', 'genres_names']].head(3))
```

```python
# Como vemos arriba, tenemos filas que al descomprimirlas nos dan listas vacías []

# Columnas con listas (_names)
columnas_listas = ['genres_names', 'production_companies_names', 'production_countries_names',
                   'keywords_names', 'cast_names', 'crew_names', 'similar_movies_names']

# Contar listas vacías por columna
print("Listas vacías por columna:\n")
for col in columnas_listas:
    vacias = df_total[col].apply(lambda x: len(x) == 0).sum()
    porcentaje = (vacias / len(df_total)) * 100
    print(f"  {col:35s}: {vacias:4d} filas vacías ({porcentaje:5.2f}%)")

# Identificar filas que contienen listas vacías

print(" Identificamos las filas con listas vacías")

# Crear máscara, True si la fila tiene al menos una lista vacía
mask_problematicas = df_total[columnas_listas].apply(
    lambda row: any(len(val) == 0 for val in row), 
    axis=1
)

filas_problematicas = mask_problematicas.sum()
print(f"Filas con AL MENOS una lista vacía: {filas_problematicas}")
print(f"Porcentaje: {(filas_problematicas/len(df_total))*100:.2f}%")
print(f"Filas que se mantendrían: {len(df_total) - filas_problematicas}")

# Ver ejemplos de filas problemáticas

print("Ejemplos de filas con listas vacías (primeras 5):")

ejemplos = df_total[mask_problematicas][['title'] + columnas_listas].head(5)
for idx, row in ejemplos.iterrows():
    print(f"{row['title']}:")
    for col in columnas_listas:
        if len(row[col]) == 0:
            print(f"  ✗ {col}: []")
    print()


print(f"\nSe eliminarán {filas_problematicas} filas")
print(f"Quedarán {len(df_total) - filas_problematicas} filas para el modelo")
```

```python
# Estudiemos si esas filas son relevantes

# Separar películas con y sin listas vacías
df_con_vacias = df_total[mask_problematicas]
df_sin_vacias = df_total[~mask_problematicas]

print("1. Comparamos con el revenue :\n")
print(f"   Películas CON listas vacías:")
print(f"     - Media revenue: ${df_con_vacias['revenue'].mean():,.0f}")
print(f"     - Mediana revenue: ${df_con_vacias['revenue'].median():,.0f}")
print(f"     - Revenue = 0: {(df_con_vacias['revenue'] == 0).sum()} películas")

print(f"\n   Películas SIN listas vacías:")
print(f"     - Media revenue: ${df_sin_vacias['revenue'].mean():,.0f}")
print(f"     - Mediana revenue: ${df_sin_vacias['revenue'].median():,.0f}")
print(f"     - Revenue = 0: {(df_sin_vacias['revenue'] == 0).sum()} películas")

print("\n2. Comparamos otras métricas :\n")
comparacion = pd.DataFrame({
    'Con vacías': [
        df_con_vacias['budget'].mean(),
        df_con_vacias['popularity'].mean(),
        df_con_vacias['vote_average'].mean(),
        df_con_vacias['vote_count'].mean(),
        df_con_vacias['runtime'].mean()
    ],
    'Sin vacías': [
        df_sin_vacias['budget'].mean(),
        df_sin_vacias['popularity'].mean(),
        df_sin_vacias['vote_average'].mean(),
        df_sin_vacias['vote_count'].mean(),
        df_sin_vacias['runtime'].mean()
    ]
}, index=['Budget', 'Popularity', 'Vote Avg', 'Vote Count', 'Runtime'])

print(comparacion)
```

```python
# Al ver que no son importantes, las eliminamos
df_total = df_total[~mask_problematicas].copy()

print(f"Shape final: {df_total.shape}")
print(f"Películas eliminadas: {filas_problematicas}")
print(f"Películas mantenidas: {len(df_total)}")
```
