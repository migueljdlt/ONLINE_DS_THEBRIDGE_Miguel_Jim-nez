## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def comprobar_dni (documento):
   if len(documento) >= 7  and len(documento) <= 8:
       return True
   else:
       return False
def letra_dni(documento):
   palabra="TRWAGMYFPDXBNJZSQVHLCKE"
   letradni=documento[:-1]
   letra_correcta = palabra[int(letradni)%23]
   return letra_correcta
documento_nacional = input("Ingrese su documento nacional: ")
validacion = comprobar_dni(documento_nacional)
if validacion:
    letra_correcta = letra_dni(documento_nacional)
    if documento_nacional[-1].upper() == letra_correcta:
        print("El documento es válido")
    else:
        print("El documento no es válido")
else:
    print("El documento no es válido")






`````
